```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
fig, (plot1, plot2) = plt.subplots(1, 2)
plot1.plot(sin(x), x)
plot1.set_title('sin(x)')
plot2.plot(cos(x), x)
plot2.set_title('cos(x)')
fig.savefig('prompt-8-fig.pdf')
```
